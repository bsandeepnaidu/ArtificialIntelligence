```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn import metrics  # for confusion matrix

from sklearn.impute import SimpleImputer 

from sklearn.tree import DecisionTreeClassifier
```

```python
colnames = ['preg', 'plas', 'pres', 'skin', 'test','mass', 'pedi','age','class']
pima_df = pd.read_csv("pima-indians-diabetes.csv", names = colnames)
```

```python
#pima_df[['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age']] = pima_df[['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age']].replace(0, np.NaN)

#pima_df.head(5)
```

```python
import pandas as pd

# Assuming pima_df is already loaded with your data 

# List of columns to convert to numeric
#columns_to_convert = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age']

# Convert specified columns to numeric, coercing errors to NaN
pima_df[colnames] = pima_df[colnames].apply(pd.to_numeric, errors='coerce')
```

```python
import seaborn as sns
sns.pairplot(pima_df , hue='class' , diag_kind = 'kde')
```

```python
# Separate features (X) and target variable (y)
X = pima_df.drop(columns=['class'])
y = pima_df['class']

# Identify and select numeric columns
numeric_cols = X.select_dtypes(include=['number']).columns
X_numeric = X[numeric_cols]

# Create an instance of SimpleImputer with strategy='median' for numeric columns
imputer = SimpleImputer(strategy='median')

# Fit and transform X_numeric to impute NaN values
X_imputed = pd.DataFrame(imputer.fit_transform(X_numeric), columns=numeric_cols)

# Plot distribution before and after imputation
plt.figure(figsize=(12, 6))
plt.suptitle('Distribution Before and After Imputation', fontsize=16)

# Plot before imputation
plt.subplot(1, 2, 1)
for col in numeric_cols:
    sns.histplot(X_numeric[col], kde=True, label=col, alpha=0.5)
plt.title('Before Imputation')
plt.legend()

# Plot after imputation
plt.subplot(1, 2, 2)
for col in numeric_cols:
    sns.histplot(X_imputed[col], kde=True, label=col, alpha=0.5)
plt.title('After Imputation')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Separate features (X) and target variable (y)
X = pima_df.drop(columns=['class'])
y = pima_df['class']

# Identify and select numeric columns
numeric_cols = X.select_dtypes(include=['number']).columns

# Create an instance of SimpleImputer with strategy='median' for numeric columns
imputer = SimpleImputer(strategy='median')

# Fit and transform X to impute NaN values
X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

# Plot distribution before and after imputation for each numeric column
plt.figure(figsize=(15, 10))

for i, col in enumerate(numeric_cols, 1):
    # Plot before imputation
    plt.subplot(2, len(numeric_cols), i)
    sns.histplot(X[col], kde=True)
    plt.title(f'{col} - Before')
    
    # Plot after imputation
    plt.subplot(2, len(numeric_cols), i + len(numeric_cols))
    sns.histplot(X_imputed[col], kde=True)
    plt.title(f'{col} - After')

plt.tight_layout()
plt.show()
```

```python
X = pima_df.drop(columns=['class']).values  # Features (all columns except 'class')
y = pima_df['class'].values  # Target variable 'class'

# Create an instance of SimpleImputer with strategy='median'
imputer = SimpleImputer(strategy='median')

# Fit and transform X to impute NaN values
transformed_X = imputer.fit_transform(X)
```

```python
pima_df['class'] = pima_df['class'].astype('category')
# MICE package
```

```python
print(pima_df.dtypes)
```

```python
print(pd.isna(y).sum())  # Check for NaN values in y
```

```python
imputer_y = SimpleImputer(strategy='most_frequent')

y_imputed = imputer_y.fit_transform(pd.DataFrame(y))

y_imputed = pd.Series(y_imputed.flatten())
```

```python
print(pd.isna(y_imputed).sum())
```

```python
test_size = 0.30 # taking 70:30 training and test set
seed = 7  # Random numbmer seeding for reapeatability of the code
X_train, X_test, y_train, y_test = train_test_split(transformed_X, y, test_size=test_size, random_state=seed)


dt_model = DecisionTreeClassifier(criterion = 'entropy' )
dt_model.fit(X_train, y_train)
```

```python
print(dt_model.score(X_train , y_train))
print(dt_model.score(  , y_test))
```

```python
y_predict = dt_model.predict(X_test)
print(metrics.confusion_matrix(y_test, y_predict))
```
